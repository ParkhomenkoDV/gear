# Исходные данные

```python
tp = "7215"
d = 75
D = 130
B = 25
Dw = 11.112 if tp[0] == "3" else None
alpha = radians(16)
e = 0.44
```

```python
if tp[0] == "3":  # шариковый
    p = 3
elif tp[0] == "7":  # роликовый
    p = 10 / 3
else:
    raise Exception

print(f"{p=}")
```

```python
# Частота вращения (об/мин)
n = 54.41
# Ресурс (ч)
resourse = 18_000
# Нагрузка (Н)
Fr_A_max, Fr_B_max, Fa_max = 19_305, 10_553, 1439
# Коэффициент режима нагружения (-)
kh = 0.85
```

```python
Cr = 130_000
C0r = 16_400
```

```python
v = 1
```

```python
safety = 1.3
safety_t = 1
```

```python
a1, a23 = 0.7, 0.21
```

# Расчет

```python
Fr_A, Fr_B, Fa = kh * Fr_A_max, kh * Fr_B_max, kh * Fa_max
print(f"{Fr_A=:.1f}")
print(f"{Fr_B=:.1f}")
print(f"{Fa  =:.1f}")
```

```python
Fe_A, Fe_B = 0.83 * e * Fr_A, 0.83 * e * Fr_B
print(f"{Fe_A=:.1f}")
print(f"{Fe_B=:.1f}")
```

```python
if Fe_B - Fa_max - Fe_A < 0:
    F_A = Fe_A
if Fe_A + Fa_max - Fe_B > 0:
    F_B = Fa_max + Fe_A
```

```python
e_A = F_A / (v * Fr_A)
e_B = F_B / (v * Fr_B)

print(f"{e_A=:.2f}")
print(f"{e_B=:.2f}")
```

```python
if e_A < e:
    X_A, Y_A = 1, 0
else:
    X_A, Y_A = 0.4, 1.6
```

```python
if e_B < e:
    X_B, Y_B = 1, 0
else:
    X_B, Y_B = 0.4, 1.6
```

```python
Pr_A = (v * X_A * Fr_A + Y_A * F_A) * safety * safety_t
Pr_B = (v * X_B * Fr_B + Y_B * F_B) * safety * safety_t

print(f"{Pr_A=:.1f}")
print(f"{Pr_B=:.1f}")
```

```python
Pr = max(Pr_A, Pr_B)
print(f"{Pr=:.1f}")
```

## Проверка по грузоподъемности

```python
C = Pr * ((60 * n * resourse) / (10**6 * a1 * a23)) ** (1 / p)
print(f"{C=}")
```

```python
assert C <= Cr
```

## Проверка по долговечности

```python
L = (a1 * a23) * (Cr / Pr) ** p * 10**6 / 60 / n
print(f"{L}")
```

```python
assert resourse / 2 < L < 2 * resourse
```

## Проверка неусталостного разрушения

```python
assert Pr <= Cr / 2
```
